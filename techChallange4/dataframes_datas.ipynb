```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import plotly.express as px
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Rogeriom49/4DTAT/main/techChallange4/datasets/dados_mes_ano.csv', sep=",")
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
df.info()
```

```python
df['quadrimestre'] = df['data'].dt.quarter
```

```python
df.head()
```

```python
b1 = [1,2,3,4,5,6]
b2 = [7,8,9,10,11,12]

def bimestre(x):
    if x in b1:
        return 1
    elif x in b2:
        return 2
```

```python
t1 = [1,2,3,4]
t2 = [5,6,7,8]
t3 = [9,10,11,12]

def trimestre(x):
    if x in t1:
        return 1    
    elif x in t2:
        return 2
    elif x in t3:
        return 3
```

```python
df['bimestre'] = df['mes'].apply(bimestre)
```

```python
df['trimestre'] = df['mes'].apply(trimestre)
```
