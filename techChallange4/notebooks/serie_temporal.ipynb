```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('datasets/dados_puros.csv', parse_dates=['data'], index_col='data')
```

```python
df.head()
```

```python
resultados = seasonal_decompose(df, period=365)
```

```python
resultados.seasonal.index
```

```python
px.line(x=resultados.seasonal.index, y=resultados.seasonal)
```

```python
px.line(x=resultados.trend.index, y=resultados.trend)
```
