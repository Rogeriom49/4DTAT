```python
dados_abertos_1 = pd.melt(df, id_vars=['Id','País'],var_name='Anos',value_name='Valor')
dados_abertos_1
```

```python
dados_abertos_2 = dados_abertos_1[dados_abertos_1['Anos'].str.contains(".1", regex=False)]
dados_abertos_2
```

```python
dados_abertos_1 = dados_abertos_1.rename(columns={'Valor':'quantidade'})
```

```python
dados_abertos_1
```

```python
dados_abertos_1 = dados_abertos_1[dados_abertos_1['Anos'].str.contains('.1', regex=False) == False]
dados_abertos_1
```

```python
dados_abertos_2['Anos'] = dados_abertos_2['Anos'].map(lambda x: str(x)[:-2])
dados_abertos_2
```

```python
display(dados_abertos_1)
display(dados_abertos_2)
```

```python
dados_abertos_1['identificador'] = dados_abertos_1['País']+'-'+dados_abertos_1['Anos']
```

```python
dados_abertos_1
```

```python
dados_abertos_2['identificador'] = dados_abertos_2['País']+'-'+dados_abertos_2['Anos']
dados_abertos_2
```

```python
dados = pd.merge(dados_abertos_1, dados_abertos_2, on = 'identificador')
dados
```

```python
dados.drop(columns=['País_y', 'Anos_y', 'Id_y'], inplace=True)
```

```python
dados = dados.rename(columns={'País_x':'pais', 'Anos_x':'anos','Valor':'valor'})
```

```python
dados
```

```python
dados_novos = dados.drop(columns=['Id_x','identificador']);
```

```python
dados_novos
```

```python
anos = dados_novos['anos'].unique().tolist()
ultimos_quinze_anos = anos[-15:]
ultimos_quinze_anos
```

```python
selecao = dados_novos['anos'].isin(ultimos_quinze_anos)
df_final = dados_novos[selecao]
```

```python
df_final
```

```python
df_final['origem'] = 'Brasil'
```

```python
df_final
```

```python
df_final['tipo'] = 'Vinhos'
```
