```python
produtos = ['VINHO DE MESA','VINHO FINO DE MESA (VIN√çFERA)','SUCO','Espumante','Espumante moscatel']
```

```python
selecao = df['produto'].isin(produtos)
df_filtrado = df[selecao]
display(df_filtrado)
```

```python
df_filtrado.replace({'produto':{'VINHO FINO DE MESA (VIN√çFERA)':'VINHO DE MESA','Espumante moscatel':'Espumante'}},inplace=True)
```

```python
df_filtrado.set_index('produto')
```

```python
df_filtrado.drop(columns=['id'],inplace=True)
```

```python
def id_por_tipo(tipo):
    if (tipo == 'VINHO DE MESA'):
        return 1
    elif (tipo == 'SUCO'):
        return 2
    elif (tipo == 'Espumante'):
        return 3
```

```python
# df_filtrado['id'] = df_filtrado['produto'].apply(id_por_tipo)
df_filtrado
```

```python
df_filtrado.replace({'produto':{'SUCO':'Suco de Uva','VINHO DE MESA':'Vinho de Mesa','Espumante':'Espumantes'}},inplace=True)
df_filtrado
```

```python
df_somado = df_filtrado.groupby(df_filtrado['produto']).sum(numeric_only=True).reset_index()
df_somado
```

```python
# df_somado.drop(columns='id',inplace=True)
df_somado
```

```python
df_melted = pd.melt(df_somado, id_vars=['produto'],var_name='ano',value_name='quantidade')
df_melted
```

```python
anos = df_melted['ano'].unique().tolist()
ultimos_quinze_anos = anos[-15:]
ultimos_quinze_anos
```

```python
selecao = df_melted['ano'].isin(ultimos_quinze_anos)
df_final = df_melted[selecao]
df_final
```

```python
df_melted.to_csv('datasets/saidas/producao/producao.csv',sep=';')
```
