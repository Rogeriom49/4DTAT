Vol (Volume) é em milhões
Var (variação) é em porcentagem

```python
df['vol'] = df['vol'].str.replace('M','',regex=False)
```

```python
df['var'] = df['var'].str.replace('%','',regex=False)
```

```python
df['vol'] = df['vol'].str.replace(',','.',regex=False)
df['var'] = df['var'].str.replace(',','.',regex=False)
```

```python
df['vol'] = df['vol'].str.replace('K','',regex=False)
```

```python
df['vol'].fillna('0',inplace=True)
```

```python
df['vol'] = df.vol.astype(float)
df['var'] = df['var'].astype(float)
```

# Rede neural

```python
x = df_new.drop(columns=['data','ultimo'])
y = df_new['ultimo']
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.70, random_state=101)
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
display(x_test.shape)
display(y_test.shape)
```

```python
nn = MLPRegressor(hidden_layer_sizes=(50, 50))
nn.fit(x_train, y_train)
```

```python
nn_preds = nn.predict(x_test)
```

```python
nn_mae = mean_absolute_error(y_test, nn_preds)
nn_mse = mean_squared_error(y_test, nn_preds)
nn_r2 = r2_score(y_test, nn_preds)

display(nn_mae)
display(nn_mse)
display(nn_r2)
```
