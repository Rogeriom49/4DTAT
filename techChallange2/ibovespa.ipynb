```python
df_old = pd.read_csv('datasets/Ibovespa14anos.csv')
df = pd.read_csv('datasets/Ibovespa4anos.csv')
display(df.head(20))
```

```python
df['dia'] = df['data'].dt.day
df['mes'] = df['data'].dt.month
df['ano'] = df['data'].dt.year
```

Vol (Volume) é em milhões
Var (variação) é em porcentagem

```python
df['vol'] = df['vol'].str.replace('M','',regex=False)
```

```python
df['var'] = df['var'].str.replace('%','',regex=False)
```

```python
df.info()
```

```python
df['vol'] = df['vol'].str.replace(',','.',regex=False)
df['var'] = df['var'].str.replace(',','.',regex=False)
```

```python
df.head()
```

```python
df['vol'] = df['vol'].str.replace('K','',regex=False)
```

```python
df['vol'].fillna('0',inplace=True)
```

```python
df['vol'] = df.vol.astype(float)
df['var'] = df['var'].astype(float)
```

Regressão Linear

```python
x = df_new.drop(columns=['data','ultimo'])
y = df_new['ultimo']
```

```python
display(x.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.8, random_state=101)
```

```python
display(x_test.shape)
display(y_test.shape)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
display(x_test.shape)
display(y_test.shape)
```

```python
display(lr.score(x_test,y_test))
```

```python
print('Coefficients:', lr.coef_)
print('Intercept:', lr.intercept_)
```

```python
lr_preds = lr.predict(x_test)
```

```python
mse = mean_squared_error(y_test,lr_preds)
mae = mean_absolute_error(y_test,lr_preds)
r2 = r2_score(y_test,lr_preds)
```

```python
display(mse)
display(mae)
display(r2)
```

Desicion Tree

```python
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
dt_preds = dt.predict(x_test)
```

```python
mse = mean_squared_error(y_test,dt_preds)
mae = mean_absolute_error(y_test,dt_preds)
r2 = r2_score(y_test,dt_preds)

display(mse)
display(mae)
display(r2)
```

Rede Neural

```python
nn = MLPRegressor(hidden_layer_sizes=(50, 50))
nn.fit(x_train, y_train)
```

```python
nn_preds = nn.predict(x_test)
```

```python
nn_mae = mean_absolute_error(y_test, nn_preds)
nn_mse = mean_squared_error(y_test, nn_preds)
nn_r2 = r2_score(y_test, nn_preds)

display(nn_mae)
display(nn_mse)
display(nn_r2)
```
